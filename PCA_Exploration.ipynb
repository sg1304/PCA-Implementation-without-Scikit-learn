# Compute mean value for each feature

```python
def mean(d,n): 
    sum = 0; 
    for i in range(n): 
        sum = sum + d[i];  
    return sum / n;  
res=data.loc[:,2:32]  
for feature_name in res.columns:
    print("mean of feature",feature_name,mean(res[feature_name],len(res)))
```

# Compute Standard deviation for each column

```python
import math 
  
# Function for calculating variance 
def variance(a, n): 
  
    # Compute mean (average of 
    # elements) 
    sum = 0
    for i in range(0 ,n): 
        sum += a[i] 
    mean = sum /n 
  
    # Compute sum squared  
    # differences with mean. 
    sqDiff = 0
    for i in range(0 ,n): 
        sqDiff += ((a[i] - mean)  * (a[i] - mean)) 
    return sqDiff / n 
  
def standardDeviation(arr, n): 
  
    return math.sqrt(variance(arr, n)) 
  
# Driver Code 
n = len(data) 
for feature_name in res.columns:
    print("Standard Deviation of feature",feature_name, 
      round(standardDeviation(res[feature_name], n), 3))
```

# Compute Max and Min values

```python
def smallest_num_in_list( list ):
    min = list[ 0 ]
    for a in list:
        if a < min:
            min = a
    return min
#for feature_name in res.columns:
    #print("Min value of column",feature_name,smallest_num_in_list(res[feature_name]))

def largest_num_in_list( list ):
    max = list[ 0 ]
    for a in list:
        if a > max:
            max = a
    return max
for feature_name in res.columns:
    print("Max and max value of feature:",feature_name,smallest_num_in_list(res[feature_name]),"and",largest_num_in_list(res[feature_name]))
```

# Compute Median

```python
print("The median of each feature is:\n",data.loc[:,2:32].median())
```

# Compute Skew and Kurtosis

```python
def moment(data,k):
    data_mean = sum(data)/len(res)
    return sum((x-data_mean)**k for x in data)/len(res)
for feature_name in res.columns:
    print("skew and kurtosis for feature:",feature_name,moment(res[feature_name], 3)/(moment(res[feature_name], 2)**1.5),"and",(moment(res[feature_name], 4)/(moment(res[feature_name], 2)**2)-3))
```

# Normalize each feature between [0,1]

```python
for feature_name in res.columns:
    max_value = res[feature_name].max()
    min_value = res[feature_name].min()
    res[feature_name] = (res[feature_name] - min_value) / (max_value - min_value)
res
```

# Calculate co-variance matrix

```python
import numpy as np
mean_vec = np.mean(res, axis=0)
cov_mat = (res - mean_vec).T.dot((res - mean_vec)) / (res.shape[0]-1)
print('Covariance matrix \n%s' %cov_mat)
```

# Trace of the Covariance matrix 

```python
trace = 0;  
cov_mat=cov_mat.values
for i in range(len(cov_mat)):  
    trace += cov_mat[i][i];  
trace
```

# Eigen Value Decomposition

```python
from numpy.linalg import eig
values,vectors=eig(cov_mat)
np.set_printoptions(formatter={'float_kind':'{:f}'.format})
```

# Orthonormal Matrix

Finding the dot product of each column in the eigenvector matrix

```python
val=[]
for i in range(len(vectors)-1):
    x=np.dot(vectors[:,i],vectors[:,i+1])
    val.append(x)
print(val)
b=max(val)
'{0:.15f}'.format(b)
```

Finding the magnitude of each column vector in the matrix

```python
def mag(x): 
    return math.sqrt(sum(i**2 for i in x))
for col in vectors.T:
    print(mag(col))
```

A set of vectors are orthonormal if every vector in the matrix has magnitude 1 and the set of vectors are mutually orthogonal and 2 vectors are orthogonal if they are perpendicular to each other. i.e. the dot product of the two vectors is zero. In our case the magnitude of each vector was very close to 1 and dot product very close to zero as the max value of all the dot products is very negligible which might have occured due to rounding off errors

# Sum of Eigen values

```python
sum(values)
```

The sum of eigen values is equal to the trace of the covariance matrix computed above. Trace of the covariance matrix is the sum of all diagonal elements which are actually the variances of each feature in the dataset. This further leads us to believe that the sum of eigen values is equal to the total variance of the dataset given.
